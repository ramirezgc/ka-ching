```python
csv_file ="../input/income_data.csv"
```

```python
income_data_df = pd.read_csv(csv_file)
income_data_df.head()
```

```python
income_data = income_data_df[["State_Name","State_ab", "Mean"]]

income_data = income_data.groupby("State_Name").mean()
income_data.reset_index
income_data
```

```python
income_states = income_data_df["State_Name"].unique()
income_states_ab = income_data_df["State_ab"].unique()
income_states_ab
```

```python
income_states_df = pd.DataFrame({
    "State_Name":income_states,
    "State_ab":income_states_ab
    
}) 
income_states_df
```

```python
auto_ins_data_new = pd.merge(auto_ins_data_df, income_states_df, left_on="state", right_on="State_Name")
auto_ins_data_new.head()

auto_ins_data_new.to_csv("../input/new_auto_ins_data.csv")
```

```python
merged_df = pd.merge(auto_ins_data_new, income_data, left_on="state", right_on="State_Name")
merged_df.head()
```
